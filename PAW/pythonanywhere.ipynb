# Выкладываем сайт

Чтобы ваш сайт смогли увидеть все, нужно его выложить. Самый простой хостинг для питоновских приложений - это **[pythonanywhere](https://www.pythonanywhere.com/)**

<img src="./img/1_paw.png"></img>

## Шаг 1. Регистрация

**Выбираем бесплатный аккаунт**

Он дает возможность выложить одно приложение с адресом "yourusername.pythonanywhere.com"

<img src="./img/2_plan.png"></img>

**Заполняем форму**

<img src="./img/3_form.png"></img>

## Шаг 2. Смотрим, что внутри

**Если все ок, то вы увидите панель управления**

<img src="./img/4_panel.png"></img>

Подтвердите адрес электронной почты и можно идти дальше

## Шаг 3. Создаем приложение

**Вкладка web должна выглядеть так**

<img src="./img/7_web.png"></img>

Нажимаем Next

<img src="./img/8_domain.png"></img>

Выбираем версию питона (какая у вас на компьютере)

Там будет путь к файлу, пока все оставляем как есть

<img src="./img/9_domain_2.png"></img>

Потом открывается такая панель, можете перейти по ссылке на сайт и там должно быть что-то вроде Hello from Flask - ура, все получилось

<img src="./img/13_main.png"></img>


## Шаг 4. Экспортируем проект с гитхаба

**Почему так лучше, чем просто загрузить файлы?**

1. Вы сможете контролировать, что происходит в проекте и не запутаться, где какой файл и что вы там наисправляли (проверено на себе, это не очень)
2. Будете настоящим разработчиком
3. Не придется по одному загружать файлы

**Bash**

На панели есть раздел Consoles, там выбираем **bash** 

<img src="./img/5_bash.png"></img>

После этого можем делать ```git clone``` и потом ```git pull``` (а изменений будет много, потому что придется что-то доделывать и поправлять) - это нужно делать так:

1. Меняете у себя локально
2. Коммит, пуш
3. Pull на платформе
4. Обновление сайта

<img src="./img/6_git.png"></img>

## Шаг 6. Управляем ~~миром~~ приложением

Заходим на страницу **web**, там будет управление приложением
<img src="./img/13_main.png"></img>

В разделе Code можно проверить папку, где находится ваше приложение, версия питона и WSGI конфиг, который нужен более глубокой системе на сервере, которая управляет приложениями на все сервере.

<img src="./img/14_web_logs.png"></img>

Чтобы прописать, где наше приложение, мы должны зайти в файл, в названии которого есть wsgi, и поменять путь к проекту и имя приложения.

```project_home = u'/home/dkbrz2/test_app' # здесь название репозитория (который вы склонировали, у меня это test_app)
...
from app import app as application  # первое app потому что app.py (может быть другое имя у вас типа flask_app, my_app или как вы там назовете
```

Сохраняем и перезапускаем (зеленой кнопкой на панели web)


<img src="./img/17_wsgi.png"></img>


Еще здесь есть ссылки на файлы, куда сервер пишет то, что там проиходит с вашим приложением (например, ошибки). Если у вас что-то не работает на сайте, заходите сюда и смотрите последнюю по времени запись

<img src="./img/15_error.png"></img>
